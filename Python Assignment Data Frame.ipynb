#### Bar Plot

```python
fig_dims = (20,9)
fig, ax = plt.subplots(figsize=fig_dims)
sns.barplot(x = "City", y = "Murder", ax=ax, data=crimedf)
plt.xticks(rotation=45,fontsize=13,color='black')
```

#### Pie Chart

```python
Values=[crimedf.Murder.sum(),crimedf.Assault.sum(),crimedf.UrbanPop.sum(),crimedf.Rape.sum()]
my_labels = ('Murder','Assault','UrbanPop','Rape')
plt.pie(Values,labels=my_labels,autopct='%1.1f%%', textprops={'fontsize': 14})
plt.title('Crime',fontsize=20)
plt.legend(fontsize=12) 
plt.axis('equal')

plt.show()
```

#### Dot Plot

```python
fig, ax = plt.subplots(figsize=(16,9),dpi= 80)
ax.hlines(y=crimedf.index, xmin=0, xmax=30, color='gray', alpha=0.7, linewidth=1, linestyles='dashdot')
ax.scatter(y=crimedf.index, x=crimedf.Murder, s=75, color='firebrick', alpha=0.7)

# Title, Label, Ticks and Ylim
ax.set_title('Dot Plot for Murder', fontdict={'size':22})
ax.set_xlabel('Murder')
ax.set_yticks(crimedf.index)
ax.set_yticklabels(crimedf.City.str.title(), fontdict={'size':11},color='black')
ax.set_xlim(0, 30)
plt.show()
```

#### 4. Use Bank Dataset from LMS

A)change all the categorical columns into numerical by creating Dummies and using label encoder.

B) rename all the column names DF

C) Rename only one specific column in DF

```python
DF=pd.read_csv('bank-full.csv',sep=";")
DF_copy=DF.copy()
DF
```

#### 4A)

```python
print(DF.info())
```

```python
from sklearn.preprocessing import LabelEncoder

lb_make = LabelEncoder()
DF['job'] = lb_make.fit_transform(DF['job'])
DF['education'] = lb_make.fit_transform(DF['education'])
DF['housing'] = lb_make.fit_transform(DF['housing'])
DF['loan'] = lb_make.fit_transform(DF['loan'])
DF['contact'] = lb_make.fit_transform(DF['contact'])
DF.head() 
```

```python
marriage_status=pd.get_dummies(DF['marital'])
marriage_status
```

```python
DF_default=pd.get_dummies(DF['default'])
DF_default
```

```python
DF_month=pd.get_dummies(DF['month'])
DF_month
```

```python
DF_poutcome=pd.get_dummies(DF['poutcome'])
DF_poutcome
```

```python
DF_y=pd.get_dummies(DF['y'])
DF_y
```

```python
DF1=DF.drop(['marital'],axis=1)
DF1
```

```python
DF2=DF1.drop(['default','month','poutcome','y'],axis=1)
DF2
```

```python
Bankdf=pd.concat([DF2,marriage_status,DF_default,DF_month,DF_poutcome,DF_y],axis=1)
Bankdf
```
